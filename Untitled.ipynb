```python
import pandas as pd
from datetime import datetime
```

```python
# Storing file names in a list
import os
file_names = []
files = [file for file in os.listdir(dir)]
```

```python
files_dict = {ind: name for (ind, name) in enumerate(files)}

for file in files:
    file = file[:-4]            # Removing the .csv from the end
    file_names.append(file)
```

```python
print(file_names)
```

```python
def normalize_date(date_str):
    try:
        # Try parsing date with different formats
        date_obj = datetime.strptime(date_str, '%d-%m-%Y')
    except ValueError:
        try:
            date_obj = datetime.strptime(date_str, '%d/%m/%Y')
        except ValueError:
            # If parsing fails, return None
            return None
    
    # Return the normalized date in the desired format
    return date_obj.strftime('%d-%m-%Y')
```

```python
for stock in file_names:
    stock_data = pd.read_csv(dir + '/' + stock + '.csv')
    print(stock_data['Date'].apply(normalize_date))
        
    
```
