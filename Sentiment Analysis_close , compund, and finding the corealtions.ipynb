```python
# Preprocessing the text
def preprocess_text(text):

    tokens = word_tokenize(text.lower())        # Tokenize the text in lowercase words (for better comparisons)

    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]    # Removing stopwords such as 'the', 'a', 'an' etc

    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]  # Lemmatization of tokens

    processed_text = ' '.join(lemmatized_tokens)    # Combining back the tokens
    return processed_text
```

```python
# Finding the sentiment of the text

analyzer = SentimentIntensityAnalyzer()

# Function for getting the sentiment in a binary value

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    sentiment = 1 if scores['pos'] > 0 else 0
    return scores['compound']

# Apply get sentiment function
df['Sentiment_New'] = df['Sentence'].apply(get_sentiment)
df
```

```python
def ProperNounExtractor(text):
    word_list = []
    sentences = nltk.sent_tokenize(text)
    for sentence in sentences:
        words = nltk.word_tokenize(sentence)
        words = [word for word in words if word not in set(stopwords.words('english'))]
        tagged = nltk.pos_tag(words)
        for (word, tag) in tagged:
            if tag == 'NNP': # If the word is a proper noun
                word_list.append(word)
    return word_list
```
